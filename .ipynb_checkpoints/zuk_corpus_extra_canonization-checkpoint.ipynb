## Combine into one DataFrame

```python
add_names = lambda df, source: df.assign(source=source)
sources = [wiki, yout, youtc, youtd, twit, goog]
source_names = ['wikipedia', 'youtube',  'youtube', 'youtube', 'twitter', 'google']
lut = pd.concat([add_names(x, name) for x,name in zip (sources, source_names)], sort=True)
lut.head()
```

## Cleaning steps
1. strip
1. unshorten (https://github.com/DocNow/unshrtn)
1. check canonical vs. long
1. youtube.com / youtu.be
1. unicode issues


### Strip

```python
CANONIZATION_RE = '\\w*:?(?://)?(?:www\.)?(.+[^/])/?$'
```

```python
lut['uri_cln'] = lut.uri.str.extract(CANONIZATION_RE)
```

### Extract domain

```python
import tldextract as tld
def tryextract(dns):
    try:
        return '.'.join(tld.extract(dns)[-2:])
    except TypeError:
        return ""
    
lut = lut.assign(tld = lut.uri.map(tryextract))
```

```python
lut.uri_cln.nunique()
```

### Unshrtn

```python
lut[lut.tld=='bit.ly'].head()
```

shortener list taken from here https://www.techmaish.com/list-of-230-free-url-shorteners-services/

```python
shrtn = [s.strip().lower() for s in open('shorteners.txt').readlines()]
shrtn.extend(['dlvr.it', 'ift.tt', 'ln.is', 'shar.es', 'j.mp', 'fb.me'])
shrtn[:5]
```

```python
lut['is_shortened'] = lut.tld.isin(shrtn)
```

```python
lut[lut.tld.isin(shrtn)].tld.value_counts()
```

```python
unshrtn_comm = 'http://localhost:3000?url=' + lut[lut.tld.isin(shrtn)].uri.drop_duplicates()
with open('unshrtn_commands.txt', 'wb') as f:
    for c in unshrtn_comm.tolist():
        f.write((c+'\n').encode('utf8'))
```

```python
import json
responses = []
with open('unshrtn_backup.jsonl', 'r', encoding='utf8') as f:
    for j in f.readlines():
        responses.append(json.loads(j.strip()))
```

```python
with open('unshrtn_backup2.jsonl', 'r', encoding='utf8') as f:
    for j in f.readlines():
        responses.append(json.loads(j.strip()))
```

```python
with open('unshrtn_backup3.jsonl', 'r', encoding='utf8') as f:
    for j in f.readlines():
        responses.append(json.loads(j.strip()))
```

```python
responses[0]
```

```python
sl = pd.DataFrame([(r[1]['short'], r[1]['long']) for r in responses], columns=['uri', 'uri_unshrtn'])
sl.head()
```

```python
lut['uri_unshrtn'] = lut.uri.map(sl.drop_duplicates().set_index('uri')['uri_unshrtn'])
```

```python
lut.head()
```

```python
lut['uri_cln_unshrtn'] = lut.uri_unshrtn.str.extract(CANONIZATION_RE)[0].combine_first(lut.uri_cln)
lut.head()
```

```python
lut.uri_unshrtn.isna().value_counts()
```

```python
lut.groupby(['source', 'is_shortened']).size()
```

### Raw unshorten

```python
import re
CLEAN_URI_RE = re.compile('([^?\']+)')
```

#### bit.ly

```python
all_bitly_raw = pd.read_csv('all_bitly_raw.csv')
```

```python
all_bitly_raw.head()
```

```python
lut['uri_unshrtn_bitly'] = lut.uri.map(all_bitly_raw[['uri', 'long_url']].drop_duplicates().set_index('uri')['long_url'])
```

```python
lut.head()
```

```python
lut['uri_cln_unshrtn'] = lut.uri_unshrtn_bitly.str.extract('\\w*:?(?://)?(.+[^/])/?$')[0].combine_first(lut.uri_cln_unshrtn)
```

#### The rest

```python
unsh_all_df = pd.read_csv('unshorten_raw_requests.csv')
```

```python
lut['uri_unshrtn_raw'] = lut.uri.map(unsh_all_df[['uri', 'long_url']].drop_duplicates().set_index('uri')['long_url'])
lut['uri_cln_unshrtn'] = lut.uri_unshrtn_raw.str.extract('\\w*:?(?://)?(.+[^/])/?$')[0].combine_first(lut.uri_cln_unshrtn)
```

### Extract unshortened domain

```python
lut = lut.assign(tld_unshrtn = lut.uri_cln_unshrtn.map(tryextract))
```

### youtube.com / youtu.be

```python
lut['uri_cln'] = lut['uri_cln'].str.replace('youtu.be', 'youtube.com')
lut['tld'] = lut['tld'].str.replace('youtu.be', 'youtube.com')
```

```python
lut['uri_cln_unshrtn'] = lut['uri_cln_unshrtn'].str.replace('youtu.be', 'youtube.com')
lut['tld_unshrtn'] = lut['tld_unshrtn'].str.replace('youtu.be', 'youtube.com')
```

## Write final dataset

```python
lut.to_csv('lang_uri_time_all.csv.gz', compression='gzip', index=False)
```

```python
tld_counts = (lut.groupby(['source','tld'])
 .uri.nunique()
 .sort_values(ascending=False)
 .unstack(level=0)
 .assign(total=lambda x: x.sum(axis=1))
 .fillna(0)
 .sort_values('total', ascending=False))

tld_counts.head(20)
```

```python
tld_ratios = tld_counts.drop('total', axis=1).div(tld_counts.drop('total', axis=1).sum(axis=0), axis=1)
tld_ratios['mean'] = tld_ratios.apply(np.mean, axis=1)
tld_ratios = tld_ratios.sort_values('mean', ascending=False).round(2)
tld_ratios.head(30)
```
